```
import pandas as pd
import numpy as np 
from tensorflow import keras
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential, Input, Model
from keras.layers import Dense, Embedding, LSTM, Bidirectional, Dropout
from sklearn.model_selection import train_test_split
import re
from spacy.lang.en.stop_words import STOP_WORDS
from sklearn.metrics import accuracy_score, f1_score
```

```
def tokenize_text(text):
    text = text.lower()
    text = re.sub('[^a-z]', ' ', text)
    text = text.split()
    text = [ word for word in text if word not in STOP_WORDS and len(word)>2 and word in embed]
    text = ' '.join(text)
    return text
```

```
VEC_DIM = 300
max_len = 36
df = pd.read_csv('sample100k.csv')
df['label'] = (df['label'] == 2).astype('int')
df = df[['title','label']]
# ta = df[df.label==0]
# fa = df[df.label==1]
# frames = [ta.sample(30000), fa]
# df = pd.concat(frames)
df = df[pd.notnull(df['title'])]
df = df.sample(frac = 1)
```

```
len(df[df.label==2])
```

```
embed = {}
def load_embed():
    with open('glove_title.txt','r') as f:
        for l in f:
            l = l.split()
            try:			
            	embed[l[0]] = np.asarray(l[1:], dtype = 'float32')
            except:
            	pass
load_embed()
```

```
n = df.shape
print(n)

df.index = range(n[0])
X  = df['title']
Y = df['label']
# Y = [0 if x==0 else 1 for x in Y]
# Y = df['label'].apply(lambda x: 0 if x==0 else 1)
X = X.apply(lambda x: tokenize_text(x))
X_train, X_test , Y_train, Y_test = train_test_split(X, Y , test_size = 0.20,stratify=Y,random_state=1)
```

```
def get_embedding_matrix(features, vec_dim):
    embedding_matrix = np.zeros((features+1, vec_dim))
    for word, index in tokenizer.word_index.items():
        try:
            embedding_matrix[index] = embed[word]
        except:
            pass
    return embedding_matrix
```

```
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X)
# tokenizer.fit_on_texts(X_train)
max_features = len(tokenizer.word_counts)
embedding_matrix = get_embedding_matrix(max_features, VEC_DIM)
print(max_features)
X_train = tokenizer.texts_to_sequences(X_train)
X_train = pad_sequences(X_train, maxlen=max_len)
X_test = tokenizer.texts_to_sequences(X_test)
X_test = pad_sequences(X_test , maxlen=max_len)
```

```
inp1 = Input(shape=(max_len,))
x = Embedding(embedding_matrix.shape[0], VEC_DIM, weights=[embedding_matrix])(inp1)
x = Bidirectional(LSTM(256, return_sequences=True))(x)
x = Dense(64, activation="relu")(x)
x = Dropout(0.1)(x)
x = Dense(64, activation="sigmoid")(x)
x = Dense(64, activation="sigmoid")(x)
x = Dense(2, activation="sigmoid")(x)    
model = Model(inputs=inp1, outputs=x)
```

```
# import livelossplot
# plot_losses = livelossplot.PlotLossesKeras()
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', )
history = model.fit(X_train,Y_train, batch_size=256, epochs=6,verbose=1 , validation_split = 0.1)
```

```

Y_pred = model.predict(X_test)
cp = [x for x in Y_pred]
# print(Y_pred)
Y_pred = [1 if sum(x)/len(x)>0.5 else 0 for x in Y_pred]
# # Y_test = Y_test.to_numpy()
# #for i in range(len(Y_pred)):
# #	print(Y_pred[i],Y_test[i])
print(accuracy_score(Y_test,Y_pred))
print(f1_score(Y_test,Y_pred,average='weighted'))
print(f1_score(Y_test,Y_pred,average='micro'))
```

```
import matplotlib.pyplot as plt
import itertools
def plot_confusion_matrix(cm, classes,normalize=False,title='Confusion matrix',cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        print(j,i,cm[i,j])
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```
from sklearn import metrics
cm = metrics.confusion_matrix(Y_test, Y_pred)
plot_confusion_matrix(cm, classes=['REAL', 'FAKE'])
```
